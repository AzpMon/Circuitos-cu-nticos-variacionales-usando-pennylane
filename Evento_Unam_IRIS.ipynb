### APRENDIZAJE DE MÁQUINA SUPERVISADO

El *aprendizaje de máquina supervisado* pertenece a una de las categorías de los problemas del aprendizaje  de máquina donde se utilizan conjuntos de datos (_data sets_) previamiente etiquetados para entrenar algoritmos que reconozcan patrones y hagan predicciones.

Asumimos que los datos son representados mediante vectores reales $\textbf{x}_n$ de dimensión $D$. Las componentes de este vector son llamados _características_ o _atributos_. El conjunto de todos estos datos está definido como

$$\{\textbf{x}_1,\dots,\textbf{x}_N\} \subset \mathbb{R}^{D}.$$

La particularidad del aprendizaje supervisado reside en que tenemos etiquetas $y_n$ asociadas a cada vector $\textbf{x}_n$ y nuestro data set se define como un conjunto de pares vector-etiqueta:

$$\{(\textbf{x}_1,y_1),\dots,(\textbf{x}_n,y_n)\}.$$

Con los datos _correctamente representados_, podemos construir un modelo que tome los datos y produzca una etiqueta, esto es, el modelo describe una relación entre las etiquetas $y_n$ y los vectores $\textbf{x}_n$ la cual de manera muy general escribimos como:

$$y_n=f(\textbf{x}_n)+\epsilon$$

donde

$$f:\mathcal{X}\rightarrow \mathcal{Y}$$

es nuestro modelo; $\mathcal{X}, \mathcal{Y}$ son los conjuntos donde viven nuestros datos y sus etiquetas, respectivamente.
El termino $ϵ$ corresponde al *error*, que considera la desviacion de las predicciones de los valores reales, y está descrito por una distribución normal con media cero.

Observe que el modelo puede tomar datos distintos de los datos que consideramos para entrenar el modelo, y producir sus correspondientes etiquetas. En general buscamos que el modelo se comporte bien, es decir, produzca predicciones precisas en datos no vistos anteriormente.

En general se construye un modelo de la forma $f(\cdot,\theta)$, con un parametro $\theta$ para el cual esperamos encontrar un buen $\theta^*$ que ajuste bien los datos con las etiquetas, es decir,

$$f(\mathbf{x}_n,\theta^*)\approx y_n, \hspace{8pt} \text{para cada} \hspace{8pt} \mathbf{x}_n\in\{(\textbf{x}_1,y_1),\dots,(\textbf{x}_n,y_n)\}.$$

#### ¿Cómo podemos encontrar los mejores parámetros para nuestro problema?

Una vez construido un modelo, se definien algunas funciones a continuacion para evaluar su rendimiento, es decir, determinar qué tan bien se ajusta al data set.

En este caso distinguimos el valor de producido por el modelo (i.e. la prediccion) como $\hat{y}_n(\theta) := f_{\theta}(\mathbf{x}_n).$

###### Función de pérdida:

Una _función de pérdida_ es una función que evalúa el error de la predicción de un dato con respecto a su correspondiente etiqueta o valor real. Es decir, si tomamos $(\mathbf{x}_i, y_i)$ de nuestro conjunto de entrenamiento, la función de pérdida $\ell$ compara la prediccion $\hat{y}_n(\theta)$ con su etiqueta $y_i$, por ejemplo

$$\ell(\hat{y}_n(\theta),y_n) = \lVert \hat{y}_n(\theta) - y_n \rVert ^2.$$

###### Funcion de costo

Una _función de costo_ es una función que evalúa el desempeño del agregado de todas las predicciones realizadas sobre un conjunto de datos. Si consideramos el conjunto de datos de entrenamiento $\{(\mathbf{x}_1, y_1), \dots, (\mathbf{x}_N, y_N)\}$, una función de costo $\mathbb{L}$ puede ser el promedio de las pérdidas, i.e. $$\mathbb{L}(\theta,b) = \frac{1}{N} \sum_{n=1}^N \ell(\hat{y}_n(\theta), y_n).$$

###### Exactitud (accuracy)

La _exactitud_ se define simplemente como la proporción de precisiones correctas con el total de predicciones realizadas: $$\text{Acc} = \frac{\text{Número de predicciones correctas}}{\text{Número total de predicciones}}$$

##### Optimizacion

A partir de estas funciones definimos un problema de optimizacion que nos permita encontrar los mejores parametros para nuestro problema. A saber, buscamos minimizar la funcion de costo. Existen distintas tecnicas para encontrar este minimo. En este trabajo usamos una funcion predefinida para resolver el problema.

### CIRCUITOS CUÁNTICOS VARIACIONALES

El aprendizaje de máquina cuántico extiende los principios del aprendizaje de máquina clásico utilizando propiedades de la mecánica cuántica como la superposición, el entrelazamiento, y particularmente la estructura del espacio donde viven los qubits, para buscar una mejora en el rendimiento del aprendizaje.

Ahora nuestros modelos corresponden a circuitos cuánticos, que de la misma forma dependen de parámetros que buscamos **optimizar de manera clásica** a través de una función de costo. En el caso de _circuitos cuánticos variacionales_ realizamos este proceso de manera iterativa, variando los parámetros y retroalimentando al modelo hasta encontrar una buena solución.

El proceso a repetirse iterativamente se realiza de la siguiente forma. Primero se codifican los datos en estados cuanticos, luego se procesan atraves del modelo, i.e. un circuito parametrizado por $\theta$, lo cual produce predicciones. Estas predicciones se comparan con la correspondiente etiqueta para producir la funcion de costo, la cual se minimiza para obtener un nuevo valor del parametro, el cual se alimenta nuevamente al modelo. En este proceso de evaluacion de la funcion de costo y optimizacion se realizan de manera clasica.

#### **Funciones cuánticas**
Un circuito cuántico es construido como una funión de python. Esta función tiene que regresar una medición, probabilidades o valores esperados.

```
import pennylane as qml
def circuito_ejemplo(parametro1, parametro2):
  """En este circuito, esperamos que se realice una serie de compuertas cuánticas sobre 2 qubits, y
  nos regrese el valor esperado para el observable PauliZ en el qubit 0"""
  qml.Hadamard(wires=0)
  qml.Hadamard(wires=1)
  qml.RY(phi = parametro1, wires=0)
  qml.RY(phi = parametro2, wires=1)
  qml.CNOT(wires=[1,0])

  return qml.expval(qml.PauliZ(wires=0))
```
Si ejecutamos el circuito tal y como esta con cualesquiera parametros, no obtendremos el resultado deseado (un número real asociado a un valor esperado). Para evaluar una función cuántica (circuito cuántico) es necesario ejecurarlo en un *dispositivo cuántico* dento de un *QNODE*

#### **Dispositivos cuánticos**
En Pennylane es necesario especificar dónde se ejecutara nuestro circuito. Cualquier objeto computacional que pueda aplicar operaciones cuánticas y devolver un valor de medición se llama dispositivo cuántico, estos pueden ser simuladores o dispositivos de harware real.
```
dispositivo_ejemplo = qml.device(name = 'default.qubit', wires=2)
```

#### **QNODES**

En PennyLane, los nodos cuánticos (QNodes) representan circuitos cuánticos. Están vinculados a un dispositivo específico que los ejecuta y evalúa valores de expectativa y varianza. Estos se pueden definir con un decorador. Según la interface que utilicemos será el tipo de dato que regrese la evaluación. Por default son arreglos de numpy pero se puede configurar para que sean tensores de pytorch, tensorflow o jax. Todo junto:

```
dispositivo_ejemplo = qml.device(name = 'default.qubit', wires=2)
@qml.qnode(device=dispositivo_ejemplo,interface = 'autograd')
def circuito_ejemplo(parametro1, parametro2):
  """En este circuito, esperamos que se realice una serie de compuertas cuánticas sobre 2 qubits, y
  nos regrese el valor esperado para el observable PauliZ en el qubit 0"""
  qml.Hadamard(wires=0)
  qml.Hadamard(wires=1)
  qml.RY(phi = parametro1, wires=0)
  qml.RY(phi = parametro2, wires=1)
  qml.CNOT(wires=[1,0])

  return qml.expval(qml.PauliZ(wires=0))
```

````python
import pennylane as qml
import torch
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from pennylane import numpy as np


# Cargar y preparar datos
X, Y = load_iris(return_X_y=True)
X = StandardScaler().fit_transform(X)

# Filtrar clases 1 y 2
mask = (Y == 1) | (Y == 2)
X_filtered = X[mask]
Y_filtered = Y[mask]

# Dividir en entrenamiento y prueba
X_train_12, X_test_12, Y_train_12, Y_test_12 = train_test_split(
    X_filtered, 
    Y_filtered, 
    train_size=0.7, 
    random_state=42
)

# Convertir a tensores
X_train_12 = torch.tensor(X_train_12, dtype=torch.float32)
X_test_12 = torch.tensor(X_test_12, dtype=torch.float32)
Y_test_12 = torch.tensor(Y_test_12, dtype=torch.float32)
Y_train_12 = torch.tensor(Y_train_12, dtype=torch.float32)


X_train = X_train_12
Y_train = Y_train_12
Y_train_12[Y_train_12==2]=-1

X_test = X_test_12
Y_test = Y_test_12
Y_test[Y_test == 2] = -1
````

## **FUNCIONES VARIAS**

````python
def predecir_etiquetas(circuito, X, parametros):
    """
    Realiza predicciones utilizando un circuito cuántico y un conjunto de parámetros.

    Args:
        circuito (function): Función del circuito cuántico que realiza la predicción. Debe tomar un tensor de datos y un tensor de parámetros como entrada.
        X (torch.tensor): Tensor de datos de entrada para la predicción, con dimensión (num_datos, num_features).
        parametros (torch.tensor): Tensor de parámetros del circuito cuántico, con dimensión (num_capas, num_parametros_por_capa). Los parámetros se utilizan para ajustar el circuito cuántico.

    Returns:
        torch.tensor: Tensor de etiquetas predichas, con dimensión (num_datos,). Las etiquetas se determinan usando la función `torch.sign` sobre el resultado de la medición del circuito cuántico.
    """
    ejecuciones_circuito = torch.tensor([circuito(x, parametros) for x in X])
    return torch.sign(ejecuciones_circuito)




def fn_costo(circuito, X, Y, params):
    """
    Calcula la función de costo (error cuadrático medio) para un conjunto de datos utilizando un circuito cuántico.

    Args:
        circuito (function): Función del circuito cuántico que realiza la predicción. Debe tomar un tensor de datos y un tensor de
                                parámetros como entrada y devolver una predicción.
        X (torch.tensor): Tensor de datos de entrada, con dimensión (num_datos, num_features). Cada fila representa un dato.
        Y (torch.tensor): Tensor de etiquetas verdaderas, con dimensión (num_datos,). Cada valor representa la etiqueta correspondiente
                            al dato en X.
        params (torch.tensor): Tensor de parámetros del circuito cuántico, con dimensión (num_capas, num_parametros_por_capa). Los
                                parámetros se utilizan para ajustar el circuito cuántico.

    Returns:
        torch.tensor: Valor del error cuadrático medio (MSE), que es una medida de la diferencia promedio entre las predicciones del
                        circuito cuántico y las etiquetas verdaderas.
    """
    squared_sum_errors = 0
    for x, y in zip(X, Y):
        squared_sum_errors += (circuito(x, params) - y) ** 2
    MSE = torch.mean(squared_sum_errors)
    return MSE



def accuracy(predicciones, etiquetas):
    """
    Calcula la precisión de las predicciones comparadas con las etiquetas verdaderas.

    Args:
        predicciones (torch.tensor): Tensor de etiquetas predichas, con dimensión (num_datos,).
        etiquetas (torch.tensor): Tensor de etiquetas verdaderas, con dimensión (num_datos,).

    Returns:
        float: Precisión de las predicciones como una fracción entre 0 y 1. Se calcula como la proporción de predicciones correctas sobre el total de datos.
    """
    clasificaciones_correctas = 0
    for etiqueta, prediccion in zip(etiquetas, predicciones):
        if etiqueta == prediccion:
            clasificaciones_correctas += 1
    return clasificaciones_correctas / torch.numel(etiquetas)
````

## **OPTIMIZACIÓN**

````python
def optimizar_evaluar(circuito, parametros, X_train, Y_train, X_test, Y_test,
                      taza_aprendizaje=0.005, epocas = 201, plotCosto = True ):
    """
    Optimiza un modelo dado y evalúa su desempeño en los conjuntos de entrenamiento y prueba.

    Args:
        circuito (function): Función que define el circuito cuántico .
        parametros (torch.Tensor): Parámetros del modelo que se optimizarán.
        X_train (torch.Tensor): Datos de entrenamiento.
        Y_train (torch.Tensor): Etiquetas de entrenamiento.
        X_test (torch.Tensor): Datos de prueba.
        Y_test (torch.Tensor): Etiquetas de prueba.
        taza_aprendizaje (float, optional): Tasa de aprendizaje para el optimizador. Por defecto es 0.008.
        epocas (int, optional): Número de épocas para el entrenamiento. Por defecto es 60.
        plotCosto (bool, optional): Indica si se debe graficar el costo durante el entrenamiento. Por defecto es True.

    Returns:
        None
    """


    # Crear un optimizador Adam con los parámetros del modelo
    optimizer = optim.Adam([parametros], lr=taza_aprendizaje)

    # Listas para almacenar el costo durante el entrenamiento y la prueba
    costo_entrenamiento, costo_prueba = [], []


    for epoch in range(epocas):

        # Calcular el costo o pérdida (que tan malo es el modelo)
        costo_train = fn_costo(circuito, X_train, Y_train, parametros)               #Datos entrenamiento
        costo_test  = fn_costo(circuito, X_test, Y_test, parametros)                 #Datos pruebaa

        # Poner a cero los gradientes del optimizador (se acumulan por defecto en pytorch)
        optimizer.zero_grad()

        # Realizar la retropropagación (con respecto a los parámetros del modelo)
        costo_train.backward()


        # Actualizar los parámetros mediante el optimizador (descenso de gradiente)
        optimizer.step()


        # Imprimir los resultados del entrenamiento cada 20 épocas
        if epoch % 20 == 0:
            test_acc  = accuracy(Y_test, predecir_etiquetas(circuito, X_test, parametros))
            train_acc = accuracy(Y_train, predecir_etiquetas(circuito, X_train, parametros))
            print(f'Época  {epoch:1d}')
            print(f'\tCosto = {costo_train.item():0.5f}    Accuracy_train = {round(test_acc,5)}    Accuracy_test = {round(test_acc,5)}', "\n",)

        # Almacenar los valores de costo para cada época
        costo_entrenamiento.append(costo_train.item())
        costo_prueba.append(costo_test.item())

    # Graficar el costo si se ha especificado
    if plotCosto:
        plt.figure(figsize=(10, 5))
        plt.plot(range(epocas), costo_entrenamiento, label='Costo de Entrenamiento')
        plt.plot(range(epocas), costo_prueba, label='Costo de Prueba')
        plt.xlabel('Épocas')
        plt.ylabel('Costo')
        plt.title('Costo durante el Entrenamiento')
        plt.legend()
        plt.grid(True)
        plt.show()

    return parametros
````

## **CIRCUITO EJEMPLO**

````python
############################################ CIRCUITO CUÁNTICO ############################################

########## CODIFICACIÓN ########### 
def codificacion(dato):
    """""
    Ejecuta un circuito cuántico de 4 qubits constituido por una codificación de los datos U(x).

    Args:
        dato (torch.tensor): Tensor asociado a los datos de entrada del modelo, con dimensión (4,).
    """
    norma = torch.norm(dato)
    for qubit in range(4):
        qml.Hadamard(wires=qubit)
        qml.RY(phi=dato[qubit], wires=qubit)

    qml.CRY(phi=-0.3385 * norma, wires=[0, 1])
    qml.CRY(phi=-0.4946 * norma, wires=[1, 2])
    qml.CRY(phi=-0.1941 * norma, wires=[2, 3])
    qml.CRY(phi=-0.7944 * norma, wires=[3, 0])


device = qml.device(name='default.qubit', wires=4)            # Definir el dispositivo cuántico

@qml.qnode(device=device, interface='torch')                  # qNode
def circuito(dato, parametros):
    """
    Ejecuta un circuito cuántico de 4 qubits constituido por una codificación de los datos, U(x), y una
    parametrización W(θ). Se considera por regla de decisión una medición en el observable PauliZ en el
    penúltimo qubit. Dicho valor se utilizará en la regla de decisión para clasificar los datos.

    Args:
        dato (torch.tensor): Tensor asociado a los datos de entrada del modelo, con dimensión (4,).
        parametros (torch.tensor): Tensor de parámetros de tamaño (num_capas, 8) a optimizarse en el circuito.

    Returns:
        torch.tensor: Valor esperado al realizar una medición en PauliZ en el penúltimo qubit.
    """
    ###################### CODIFICACIÓN ######################
    codificacion(dato)

    #################### PARAMETRIZACIÓN ####################
    for parametro in parametros:
        for qubit in range(4):
            qml.RX(phi = parametro[qubit], wires=qubit)
        qml.CRY(phi=parametro[4], wires=[0,1])
        qml.CRY(phi=parametro[5], wires=[1,2])
        qml.CRY(phi=parametro[6], wires=[2,3])
        qml.CRY(phi=parametro[7], wires=[3,0])

    ###################### MEDICIÓN  ######################
    return qml.expval(qml.PauliZ(wires=2))



############################################ CREAR PARÁMETROS ############################################
def crear_parametros(capas):
    """
    Crea un tensor de parámetros aleatorios para el circuito cuántico. Cada parámetro se muestrea de una distribución normal.
    Args:
        capas (int): Número de capas en el circuito, determina el primer tamaño de la dimensión del tensor.

    Returns:
        torch.tensor: Tensor de parámetros de tamaño (capas, 4, 3).
    """
    return torch.randn(capas, 8, requires_grad=True)
````

#### **Implementación clásica: máquina de soporte vectorial clásica con optimización de hiperparámetros**

````python
def objective(trial):
    """Función objetivo para la optimización de hiperparámetros con Optuna."""

    # Definir el espacio de búsqueda para los hiperparámetros
    C = trial.suggest_float('C', 1e-3, 1e3, log=True)           # Parámetro de regularización
    gamma = trial.suggest_float('gamma', 1e-3, 1e3, log=True)   # Parámetro de la función de kernel
    kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'poly'])  # Tipo de kernel

    # Crear y ajustar el modelo SVM
    model = SVC(C=C, gamma=gamma, kernel=kernel)
    model.fit(X_train, Y_train)

    # Realizar predicciones
    y_pred = model.predict(X_test)

    # Calcular la precisión
    accuracy = accuracy_score(Y_test, y_pred)

    return accuracy

# Crear un objeto de estudio de Optuna para optimizar la función objetivo
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=80)

# Crear y ajustar el modelo SVM con los mejores hiperparámetros encontrados
best_params = study.best_params
best_model = SVC(C=best_params['C'], gamma=best_params['gamma'])
best_model.fit(X_train, Y_train)

# Evaluar el modelo con los mejores hiperparámetros en el conjunto de prueba
best_y_pred = best_model.predict(X_test)
best_accuracy = accuracy_score(Y_test, best_y_pred)
````

````python
# Imprimir los mejores hiperparámetros y la mejor precisión encontrada
print(f'Mejores hiperparámetros:{study.best_params}')
print(f'Mejor Accuracy_train={round(study.best_value,5)}, Mejor Accuracy_test = {round(best_accuracy, 5)}')
````
